## Aufgabe 1: Sonnenstrahlen optimal auf das Rohr spiegeln

Ein bewährtes Konzept in der mathematischen Modellierung ist es, zunächst eine (stark) vereinfachte Situtation zu betrachten und diese Schritt für Schritt an die Realität anzupassen. Wir folgen diesem Prinzip und betrachten zunächst den einfachsten Fall: Ein Fresnel-Kraftwerk mit nur einem einzigen Spiegel mittig unter dem Absorberrohr. Diese Situation ist in der obigen Abbildung dargestellt. Dabei steht
- $\gamma$ (Gamma) für den **Neigungswinkel des Spiegels**, das ist der Winkel zwischen der Spiegelnormalen (im Bild grau dargestellt) und dem Boden.
- $\alpha$ (Alpha) für den **Sonnenenifallswinkel**, das ist der Winkel zwischen einfallendem Sonnenstrahl (im Bild rot dargestellt) und Boden.

Damit der reflektierte Sonnenstrahl (im Bild grün dargestellt) genau auf das Absorberrohr trifft, muss der Neigungswinkel des Spiegel stets an den Sonneneinfallswinkel angepasst sein. Weil sich der Sonnenstand im Verlaufe des Tages ändert, muss auch der Spiegel während des Tages gedreht werde.
### Teil a)
In obiger Abbildung ist der Neigungswinkel $\gamma$ noch nicht korrekt eingestellt, sodass der reflektierte Strahl nicht das Rohr trifft. Im Gegensatz dazu zeigt die untere Abbildung, wie der reflektierte Strahl zum einfallenden Strahl stehen muss, damit er genau auf das Absorberrohr trifft.

Jedoch fehlt in deieser Abbildung der Spiegel. Wie muss der Spiegel eingezeichnet werden damit die dargestellte Reflexion **physikalisch korrekt** ist? 

*Übertrage die letzte Zeichnung auf ein Blatt. Ergänze deine Zeichnung anschließend um den Spiegel und seine Normale, sodass eine physikalisch korrekte Reflektion entsteht. Zeichne auch den Neigungswinkel $\gamma$ ein.*

### Teil b)
Nehmen wir nun an, dass der Sonneneinfallswinkel $\alpha$ gegeben ist. Stelle eine Formel zur Berechnung des korrekten Neigungswinkels $\gamma$ in Abhängigkeit von $\alpha$ auf. Dabei empfiehlt es sich, die von dir angefertigte Zeichnung zu Hilfe zu nehmen. Die Winkel müssen im **Bogenmaß** (nicht in Grad) angegeben werden. Zur Erinnerung: $360° = 2\pi$.

*Ersetze im folgenden Code `NaN` durch eine Formel in $\alpha$, welche den gesuchten Neigungswinkel $\gamma$ berechnet. Die Zahl $\pi$ wird einfach als `pi` eingegeben. Drücke anschließend auf den "Run" Button.*

**Hinweis: Bei einer Formel müssen alle Variablen hinter dem @-Zeichen benutzt werden. Sie werden beim Aufrufen der Funktion übergeben. Alle Variablen, die nicht hinter dem @-Zeichen aufgeführt sind, werden als Konstanten betrachtet.**

**Falls du Hilfe benötigst, kannst du dir hier [Hilfekarte 1](./SolarkraftwerkHilfekarte1.ipynb) ansehen.**

```octave
% Aufstellen der Berechnungsformel
berechne_gamma = @(alpha) (alpha + pi/2)/2;% Ersetze NaN durch eine Formel in alpha, welche den korrekten Winkel gamma berechnet
```
