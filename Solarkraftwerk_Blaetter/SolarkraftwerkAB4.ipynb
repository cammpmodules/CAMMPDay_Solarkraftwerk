## Aufgabe 4: Verschiebung des Spiegels
### Teil a)

Obiges Bild zeigt schematisch die Situation mit einem verschobenen Spiegel. Die Variable $x$ gibt die **Position des Spiegelmittelpunktes** relativ zum Nullpunkt an. Der Nullpunkt befindet sich senkrecht unter dem Rohr und ist eingezeichnet. Dementsprechned ergibt eine Verschiebung des Spiegels nach links negativen Werten für $x$ und eine Verschiebung des Spiegels nach rechts positive Werte für $x$. Mithilfe der obigen Abbildung sollst du nun eine Formel für den Winkel $\beta$ zwischen Boden und reflektiertem Sonnenstrahl herleiten. Auf den vorherigen Arbeitsblättern galt stets $\beta = \frac{\pi}{2}$, aber in der neuen Situation ist $\beta$ von der Position des Spiegels $x$ abhängig. Zur Berechnung von $\beta$ benötigst du außerdem die **Höhe des Absorberrohrs** $h$ über dem Boden, welche im Bild in Blau eingezeichnet ist und als bekannte Größe betrachtet werden kann. 

Bedenke, dass man bei der Berechnung von $\beta$ drei Fälle unterscheiden muss:
- $x < 0$ (dann gilt $\beta < \frac{\pi}{2}$)
- $x = 0$
- $x > 0$ (dann gilt $\beta > \frac{\pi}{2}$; mache dir das ggf. an einer eigenen Zeichung klar)

Der unten stehende Code ist mit dem Gerüst einer *if-else-Anweisung* bereits soweit vorbereitet, dass du die Formeln für die verschiedene Fälle nur noch an den richtigen Stellen eintragen musst. Dadurch wird die Funktion `berechne_beta` definiert, die $\beta$ für gegebenes $x$ und $h$ berechnet. Die Einheit von $\beta$ soll **das Bogenmaß** (nicht Grad) sein.

*Ersetze das erste `NaN` im folgenden Code durch eine Formel in `x` und `h`, die den Winkel `beta` im Fall $x < 0$ berechnet. Ersetze das zweite `NaN` durch eine Formel in `x` und `h`, die den Winkel `beta` im Fall $x > 0$ berechnet. Ersetze das letzte `NaN` durch einen Ausdruck für den Winkel `beta` im Fall $x = 0$. Die Zahl $\pi$ kann wieder als `pi` eingegeben werden. Drücke anschließend auf den "Run" Button. Falls du einen Fehler gemacht hast, wird dir das mitgeteilt. Korrigiere dich in diesem Fall.*

**Hinweis: Für die Berechnung von** $\beta$ **wird der Tangens bzw. der Arkustangens benötigt. Falls du diese Funktionen noch nicht kennst, kannst du dir hier ansehen, wie man den Tangens (analog zu Sinus und Kosinus) für Berechnungen im rechtwinkligen Dreieck nutzen kann:**

**Die Arkustangensfunktion ist die Umkehrfunktion der (eingeschränkten) Tangensfunktion.** 


```julia
# Berechnung von Beta
function berechne_beta(x,h)
    if (x < 0)
        beta = atan(h/(-x));# Ersetze NaN durch eine Formel in x, welche den Winkel beta im Fall x<0 (Verschiebung nach links) berechnet
    elseif (x > 0)
        beta = pi-atan(h/x);# Ersetze NaN durch eine Formel in x, welche den Winkel beta im Fall x>0 (Verschiebung nach rechts) berechnet
    elseif (x == 0)
        beta = pi/2;# Ersetze NaN durch einen Ausdruck für den Winkel beta im Fall x=0
    end
    return beta;
end

# Prüfung der Funktion (ab hier nichts mehr ändern)
Solarkraftwerk.pruefe_beta(berechne_beta);
```

```julia
# Position des Spiegels
x = -2;# Ersetze NaN durch eine beliebige Spiegelposition in Metern
# Höhe des Absorberrohrs
h = 4;# Ersetze NaN durch eine beliebige Höhe in Metern (diese muss natürlich positiv sein)

# Berechnung von beta mit der von dir definierten Funktion
beta = berechne_beta(x,h);# nichts ändern

# Ausgabe (hier nichts ändern)
Solarkraftwerk.outbeta(x,h,beta);
```
