## Aufgabe 3: Die Leistung am Absorberrohr

$$
P_M = P_S \cdot e

### Teil e)
Jetzt wollen wir die Uhrzeit bestimmen, zu der unser Kraftwerk die meiste Energie erzeugt, also den Zeitpunkt, zu dem die größte Leistung am Absorberrohr generiert wird. Dazu müssen wir uns zunächst klar machen, dass $P_R$ alleine von der Zeit abhängt:
- Erstens ist $P_M$ eine Funktion der Zeit, wie wir ganz am Ende von Arbeitsblatt 2 gesehen haben. 
- Zweitens ist $A$ eine Funktion von $e$ und $b$. Da die Breite $b$ des Sekundärreflektors fix ist, ist $A$ eigentlich nur von $e$ abhängig. $e$ hängt wiederum von $\alpha$ und $\gamma$ ab.
```julia
# Diese Funktionen stehen dir zur Verfügung
grad2rad(alpha)= alpha/180*pi;# nichts ändern
berechne_alpha(t) = 14.2076 + 15.221 * (t-8);# nichts ändern
berechne_P_S(t) = -3580.0058 + 696.1063 * t - 26.8157 * t^2;# nichts ändern
berechne_gamma(alpha) = (alpha + pi/2)/2;# nichts ändern
berechne_e(s,alpha,gamma) =  s * cos(gamma - alpha);# nichts ändern
berechne_P_M(P_S, alpha) = P_S * berechne_e(1.1,alpha,berechne_gamma(alpha));# nichts ändern
berechne_P_M_t(t) = berechne_P_M(berechne_P_S(t),grad2rad(berechne_alpha(t)));# nichts ändern

# Definition der festen Größen s und b
s = 1.1;# Ersetze NaN durch die Breite des Spiegels in Metern
b = 0.8;# Ersetze NaN durch die Breite des Sekundärreflektors in Metern

# Funktion (in t) zur Berechnung von e
berechne_e_t(t) = berechne_e(s,grad2rad(berechne_alpha(t)),berechne_gamma(grad2rad(berechne_alpha(t))));# Ersetze NaN durch eine Formel, die e in Abhängigkeit von t berrechnet. Du solltest dabei die Funktion berechne_e verwenden.
# Funktion (in t) zur Berechnung von A
berechne_A_t(t) = berechne_A(berechne_e_t(t),b);# Ersetze NaN durch eine Formel, die A in Abhängigkeit der Zeit berechnet. Du solltest dabei die Funktion berechne_A verwenden.
# Funktion (in t) zur Berechnung von P_R
berechne_P_R_t(t) = berechne_P_R(berechne_P_M_t(t), berechne_A_t(t));# Ersetze NaN durch eine Formel, die P_R in Abhängigkeit der Zeit berechnet. Du solltest dabei die Funktion berechne_P_R benutzen.

# Prüfung der Eingabe (ab hier nichts ändern)
Solarkraftwerk.pruefe_PRt(berechne_P_R_t)
```
