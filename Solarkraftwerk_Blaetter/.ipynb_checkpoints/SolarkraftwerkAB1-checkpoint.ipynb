## Aufgabe 1: Die Sonnenstrahlen optimal auf das Rohr spiegeln

Ein bewährtes Konzept in der mathematischen Modellierung ist es, zunächst eine (stark) vereinfachte Situtation zu betrachten und diese Schritt für Schritt an die Realität anzupassen. Wir folgen diesem Prinzip und betrachten zunächst den einfachsten Fall: Ein Fresnel-Kraftwerk mit nur einem einzigen Spiegel mittig unter dem Absorberrohr. Diese Situation ist in der obigen Abbildung dargestellt. Dabei steht
- $\gamma$ (Gamma) für den **Neigungswinkel des Spiegels**, das ist der Winkel zwischen der Spiegelnormalen (im Bild grau dargestellt) und dem Boden.
- $\alpha$ (Alpha) für den **Sonneneinfallswinkel**, das ist der Winkel zwischen einfallendem Sonnenstrahl (im Bild rot dargestellt) und Boden.

Damit der reflektierte Sonnenstrahl (im Bild grün dargestellt) genau auf das Absorberrohr trifft, muss der Neigungswinkel des Spiegel stets an den Sonneneinfallswinkel angepasst sein. Weil sich der Sonnenstand im Verlaufe des Tages ändert, muss auch der Spiegel während des Tages gedreht werde.
### Teil a)
In obiger Abbildung ist der Neigungswinkel $\gamma$ noch nicht korrekt eingestellt, sodass der reflektierte Strahl nicht das Rohr trifft. Im Gegensatz dazu zeigt die untere Abbildung, wie der reflektierte Strahl zum einfallenden Strahl stehen muss, damit er genau auf das Absorberrohr trifft.

Jedoch fehlt in deieser Abbildung der Spiegel. Wie muss der Spiegel eingezeichnet werden damit die dargestellte Reflexion **physikalisch korrekt** ist?

### Teil b)
Nehmen wir nun an, dass der Sonneneinfallswinkel $\alpha$ gegeben ist. Da die Sonne nicht durch den Boden scheinen kann, liegt $\alpha$ zwischen 0 und $\pi$. Wie muss der Winkel $\gamma$ gewählt werden, damit der reflektierte Strahl aufs Absorberrohr trifft? Stelle eine Formel zur Berechnung des korrekten Neigungswinkels $\gamma$ in Abhängigkeit von $\alpha$ auf. Dabei empfiehlt es sich, die von dir angefertigte Zeichnung zu Hilfe zu nehmen. Berücksichtige aber auch den Fall, dass die Sonne aus Westen kommt, also $\alpha \in (\frac{\pi}{2}, \pi)$ gilt. Alle Winkel müssen im **Bogenmaß** (nicht in Grad) angegeben werden. Zur Erinnerung: $360° = 2\pi$.

*Ersetze im folgenden Code `NaN` durch eine Formel in $\alpha$, welche den gesuchten Neigungswinkel $\gamma$ berechnet. Die Zahl $\pi$ wird einfach als `pi` eingegeben. Drücke anschließend auf den "Run" Button.*

**Hinweis: Bei einer Funktion müssen alle Variablen, die hinter dem Funktionsnamen in Klammern stehen, benutzt werden. Sie werden beim Aufrufen der Funktion übergeben. Alle anderen Variablen werden als Konstanten betrachtet.**

**Falls du Hilfe benötigst, kannst du dir hier [Hilfekarte 1](./SolarkraftwerkHilfekarte1.ipynb) ansehen.**

```julia
# Berechnung von Gamma
berechne_gamma(alpha) = (alpha + pi/2)/2;# Ersetze NaN durch eine Formel in alpha, welche den korrekten Winkel gamma berechnet
```

### Teil d) 
Da es schwierig ist, den Sonnenstand permanent zu messen, ist es sinnvoll, die Ausrichting des Spiegels von der aktuellen Tageszeit abhängig zu machen. Dazu muss man nur wissen, wie der Sonneneinfallswinkel $\alpha$ von der Uhrzeit $t \in [0,24)$ abhängt. An einem Tag, an dem die Sonne um etwa 7 Uhr aufgeht und um circa 19 Uhr untergeht, ist dieser Zusammenhang (für unsere Zwecke hinreichend genau) gegeben durch
$$
\alpha(t) = 14.2076 + 15.221 * (t-8).
$$
Achtung, hierbei wird $\alpha$ in **Grad** angegeben und die Zeit $t$ in **Stunden nach Tagesbeginn**. Beispielsweise entspricht 11:30 Uhr der Zeit $t= 11,5$. Du sollst nun die obige Formel in Julia implementieren und anschließend mit deren Hilfe eine Funktion aufstellen, die den benötigten Neigungswinkel $\gamma$ in Abhängigkeit der Zeit $t$ berechnet. 

*Ersetze im folgenden Code das erste `NaN` durch eine Funktion in `t`, die den Sonneneinfallswinkel nach obiger Formel berechnet. Ersetze das zweite `NaN`, durch eine Formel, die einen in Grad gegebenen Winkel `alpha` ins Bogenmaß umrechnet. Ersetze das dritte `NaN`, durch eine Formel, die den korrekten Neigungswinkel `gamma` in Abhängigkeit der Zeit `t` berechnet. In dieser Formel kannst (und solltest) du die zwei zuvor definierten Funktionen verwenden. Das letzte `NaN` ersetzt du durch einen Tageszeit zwischen ca. 7 und 19 Uhr. Drücke anschließend auf den "Run" Button. Es erscheint eine Abbildung, die den Strahlenverlauf zu der von dir gewählten Tageszeit zeigt, basierend auf der von dir berechneten Ausrichtung des Spiegels. Falls der reflektierte Strahl nicht auf das Absorberrohr trifft, bekommst du eine Fehlermeldung und musst deine Eingaben korrigieren.*

```julia
# Berechne alpha in Abhängigkeit von t
berechne_alpha(t) = 14.2076 + 15.221 * (t-8);# Ersetze NaN durch eine Funktion in t, welche den Sonneneinfallswinkel alpha nach obiger Formel berechnet.
# Umrechnung von Grad in Rad
grad2rad(alpha)= alpha/180*pi;# Ersetze NaN durch eine Formel, die den Winkel alpha von der Einheit Grad ins Bogenmaß umrechne`.
# Berechne gamma in Abhängigkeit der Zeit
berechne_gamma_t(t) = (grad2rad(berechne_alpha(t)) + pi/2)/2;# Ersetze NaN durch eine Formel in t, welche den korrekten Winkel gamma (im Bogenmaß!) berechnet. Du kannst die Funktionen `berechne_alpha` und `grad2rad` dabei verwenden.

# Tageszeit
t = 16;# Ersetze NaN durch eine Zeit zwischen 7.07 und 18.89 Stunden nach Tagesbeginn

# Abbildung und Feedback (ab hier nichts ändern)
gamma = berechne_gamma_t(t);
Solarkraftwerk.Video(t, gamma)
```
