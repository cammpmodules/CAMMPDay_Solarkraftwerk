## Aufgabe 2: Die am Spiegel generierte Leistung
Mit **Leistung der Sonnenstrahlung** $P_S$ bezeichnen wir im Folgenden die Leistung, die erbracht wird, wenn die Sonnenstrahlen **senkrecht** auf eine Fläche von einem Quadratmeter treffen. Treffen die Strahlen nicht senkrecht auf, so wird nur eine geringere Leistung erbracht. Es gilt: Je spitzer bzw. stumpfer der Einfallswinkel, desto geringer die Leistung. Der Winkel, in dem die Sonnenstrahlen auf den Spiegel des Kraftwerks treffen, ändert sich im Laufe des Tages. Die **Leistung am Spiegel** $P_M$ ist somit ebenfalls von der Tageszeit abhängig. Um die Leistung $P_M$ unter Berücksichtigung des Einfallswinkels $\alpha$ berechnen zu können, betrachten wir eine imaginäre Hilfsfläche. Diese nimmt die gleiche Leistung durch Sonnenstrahlung auf, wie unser Spiegel, sie steht jedoch senkrecht auf den Sonnenstrahlen. 

In obiger Abbildung ist diese Fläche in rot eingezeichnet. Sie wir durch die zwei äußeren Sonnenstrahlen, die auf den Spiegel treffen, begrenzt. Ihre **Breite** bezeichnen wir mit $e$. Die **Breite des Spiegels** bezeichnen wir mit $s$. Diese ist bekannt und ebenfalls in obigem Bild eingezeichnet (in schwarz). 

### Teil a)
Stelle eine Formel für die Breite der Hilfsfläche $e$ unter Verwendung der Spiegelbreite $s$ und den aus Aufgabenblatt 1 bekannten Winkeln $(\alpha, \ \gamma)$ auf. Dazu benötigst du die Definitionen von Sinus und Kosinus im rechtwinkligen Dreieck:

Nehme zunächst an, dass $\alpha$, wie in den Abbildungen dargestellt, ein spitzer Winkel ist, also im Intervall $\left(0,\frac{\pi}{2}\right)$ liegt. Überlege dir anschließend, ob deine Formel auch für stumpfe Winkel $\alpha \in \left(\frac{\pi}{2}, \pi \right)$ gültig bleibt.
 
*Ersetze im folgenden Code `NaN` durch eine Formel in $s$, $\alpha$ und $\gamma$, welche die Breite der Hilfsfläche $e$ berechnet. Drücke anschließend auf den "Run" Button. Falls deine Formel falsch ist, bekommst du eine Fehlermeldung. Korrigiere dich in diesem Fall.*

**Falls du Hilfe benötigst, kannst du dir hier [Hilfekarte 2](./SolarkraftwerkHilfekarte2.ipynb) ansehen.**

```julia
# Berechnung der Breite e
berechne_e(s,alpha,gamma) =  s * cos(gamma - alpha);# Ersetze NaN durch eine Formel in s, alpha und gamma, welche die Breite e berechnet

# Prüfung der Formel (ab hier nichts mehr ändern)
Solarkraftwerk.pruefe_e(berechne_e)
```

### Teil b)
Nun wollen wir die in Teil a) gefundene Formel nutzen, um drei konkrete Werte für $e$ zu berechnen. Dazu nehmen wir an, dass unser Spiegel eine Breite von $s = 1.1$ Metern hat. Berechne die Breite der imaginären Fläche $e$ jeweils für einen Einfallswinkel $\alpha$ von
- 20 Grad

indem du die Funktion `berechne_e` mit den richtigen Argumenten aufrufst. Denke daran, dass `berechne_e` nur mit Winkeln im Bogenmaß funktioniert. Die Funktionen `berechne_gamma` und `grad2rad`, die du bereits vom ersten Arbeitsblatt kennst, stehen dir hier wieder zur Verfügung (du musst sie nicht mehr definieren).

*Ersetze das erste `NaN` durch ein passendes Argument, sodass `e1` der Breite der Hilfsfläche bei einem Sonneneinfallswinkel von 20 Grad entspricht. Ersetze das dritte `NaN` durch ein passendes Argument, sodass `e3` der Breite der Hilfsfläche bei einem Sonneneinfallswinkel von 130 Grad entspricht. Drücke danach auf den "Run" Button.
```julia
# Diese Funktionen vom 1. Aufgabenblatt kannst du verwenden
berechne_gamma(alpha) = (alpha + pi/2)/2;# nichts ändern
grad2rad(alpha)= alpha/180*pi;# nichts ändern

# Berechne e für alpha = 20 Grad
e1 = berechne_e(1.1,grad2rad(20),berechne_gamma(grad2rad(20)));# Ersetze NaN durch das richtige Argument, sodass e1 die Breite der Hilfsfläche bei alpha = 20 Grad ist
# Berechne e für alpha = 90 Grad
e2 = berechne_e(1.1,grad2rad(90),berechne_gamma(grad2rad(90)));# Ersetze NaN durch das richtige Argument, sodass e2 die Breite der Hilfsfläche bei alpha = 90 Grad ist
# Berechne e für alpha = 130 Grad
e3 = berechne_e(1.1,grad2rad(130),berechne_gamma(grad2rad(130)));# Ersetze NaN durch das richtige Argument, sodass e1 die Breite der Hilfsfläche bei alpha = 130 Grad ist
```
