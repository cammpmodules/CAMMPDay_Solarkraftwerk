## Aufgabe 3: Die Leistung am Absorberrohr 
Wir nehmen an, dass die Breite des Sekundärreflektors $b$ (oben in Grün eingezeichnet) gegeben ist und betrachten wieder eine Spiegelfläche der Länge 1 Metern. Vom zweiten Arbeitsblatt wissen wir, dass die Leistung am Spiegel $P_M$ durch
$$
P_M = P_S \cdot e
$$
gegeben ist, wobei $P_S$ die Leistung der Sonne und $e$ die Breite der verwendeten Hilfsfläche ist. Die Größe $e$ finden wir auch im obigen Bild wieder. Aufgrund der Symmetrie (Einfallswinkel = Ausfallswinkel) entspricht sie ebenfalls der Breite des Sonnenstrahlenbündels, das vom Spiegel senkrecht nach oben reflektiert wird. 
### Teil a)
Wir gehen zunächst davon aus, dass der Sekundärreflektor nicht breiter als das Sonnenstrahlbündel ist ($e \geq b$). In diesem Fall trifft nur ein Teil der reflektierten Strahlung den Sekundärreflektor. Folglich kommt auch nur ein Teil der Leistung am Spiegel $P_M$ am Rohr an. Der genaue **prozentuale Anteil** $A$ hängt dabei natürlich von der Breite $b$ ab und soll nun von dir bestimmt werden.

*Ersetze im folgenden Code `NaN` durch eine Funktion in `e` und `b`, welche den prozentualen Anteil der Leistung $P_M$ bestimmt, der am Absorberrohr ankommt. Drücke anschließend auf den "Run" Button. Du bekommst eine Feedback zu deiner Eingabe. Falls du einen Fehler gemacht hast, korrigiere dich.*

```julia
# Funktion zur Berechnung des prozentualen Anteils
berechne_A0(e,b) = b/e*100;# Ersetze NaN durch eine Formel in e und b, die den prozentualen Anteil A von P_M berechnet, der am Rohr ankommt

# Prüfung der Funktion (ab hier nichts mehr ändern)
Solarkraftwerk.pruefe_A0(berechne_A0)
```

### Teil b)
Definiere jetzt eine Funktion, die den prozentualen Anteil $A$ auch in dem Fall, dass der Sekundärreflektor breiter ist als das Strahlenbündel ($b > e$), richtig berechnet. Der Fall $b \leq e$ soll von der Funktion auch weiter abgedeckt sein.

*Ersetze im folgenden Code `NaN` durch eine Funktion in `e` und `b`, die den prozentualen Anteil der Leistung $P_M$, der am Absorberrohr ankommt, auch im Fall $b>e$ korrekt berechnet. Drücke anschließend auf den "Run" Button. Du bekommst eine Feedback zu deiner Eingabe. Falls du einen Fehler gemacht hast, korrigiere dich.*

**Hinweis: Um die kleinere von zwei Zahlen `x` und `y` zu finden, kannst du in Julia die Funktion `min(x,y)` benutzen. Beispielsweise gilt `min(5,9) = 5` oder `min(8,-3) = -3`.**

```julia
# Funktion zur Berechnung des prozentualen Anteils
berechne_A(e,b) = min(b/e,1)*100;# Ersetze NaN durch eine Formel in e und b, die auch im Fall  b > e den prozentualen Anteil A von P_M berechnet, der am Rohr ankommt

# Prüfung der Funktion (ab hier nichts mehr ändern)
Solarkraftwerk.pruefe_A(berechne_A)
```

### Teil c)
Der prozentuale Anteil von $P_M$, der am Absorberrohr ankommt, hängt vom Sonneneinfallswinkel $\alpha$ ab. Denn $e$ hängt von $\alpha$ ab, wohingegen $b$ fix ist. Du sollst nun für drei verschiedene Einfallswinkel $\alpha$ von 
- 35 Grad
- 95 Grad 
- und 175 Grad

diesen Anteil $A$ konkret berechnen. Dazu gehen wir davon aus, dass die Breite des Sekundärreflektors durch $b = 0,8$ Meter gegeben ist. Weiterhin gilt für die Spiegelbreite $s = 1,1$ Meter. Berechne die gesuchten Anteile, indem du die von dir definierte Funktion `berechne_A` mit den richtigen Argumenten aufrufst. Die Funktionen `grad2rad`, `berechne_gamma` und `berechne_e`, die du bereits vom ersten bzw. zweiten Arbeitsblatt kennst, stehen dir hier wieder zur Verfügung (du musst sie nicht noch einmal implementieren).

*Ersetze das erste `NaN` im folgendne Code durch die Breite des Spiegels. Ersetze das zweite `NaN` durch die Breite des Sekundärreflektors. Ersetze alle weiteren `NaN` so, dass `A1` dem gesuchten prozentualen Anteil bei einem Sonneneinfallswinkel von 35 Grad entspricht, `A2` dem gesuchten prozentualen Anteil bei einem Sonneneinfallswinkel von 95 Grad entspricht und `A3` dem gesuchten prozentualen Anteil bei einem Sonneneinfallswinkel von 175 Grad entspricht. Die Kommentare am Ende der jeweiligen Zeile helfen dir dabei. Drücke danach auf den "Run" Button. Wenn du alles richtig gemacht hast, werden dir die drei berechneten Anteile ausgegeben. Falls du einen Fehler gemacht hast, bekommst du eine Fehlermeldung und musst deine Eingabe korrigieren.*

```julia
# Diese Funktionen von den vorherigen Arbeitsblättern kannst du verwenden
grad2rad(alpha)= alpha/180*pi;# nichts ändern
berechne_gamma(alpha) = (alpha + pi/2)/2;# nichts ändern
berechne_e(s,alpha,gamma) =  s * cos(gamma - alpha);# nichts ändern

# Breite des Spiegels
s = 1.1;# Ersetze NaN durch die Breite des Spiegels in Metern
b = 0.8;# Ersetze NaN durch die Breite des Sekundärreflektors in Metern

# Berechne A für alpha = 35 Grad
alpha1 = grad2rad(35);# Ersetze NaN durch einen Ausdruck für den Einfallswinkel (45 Grad) im Bogenmaß
gamma1 = berechne_gamma(alpha1);# Ersetze NaN durch einen Ausdruck für den zugehörigen Neigungswinkel
e1 = berechne_e(s,alpha1,gamma1);# Ersetze NaN durch einen Ausdruck für die zugehörige Breite der Hilfsfläche e
A1 = berechne_A(e1,b);# Ersetze NaN durch das richtige Argument, sodass A1 der Anteil der Leistung P_M ist, der bei einem Einfallswinkel von 35 Grad am Absorberrohr ankommt

# Berechne A für alpha = 95 Grad
alpha2 = grad2rad(95);# Ersetze NaN durch einen Ausdruck für den Einfallswinkel (95 Grad) im Bogenmaß
gamma2 = berechne_gamma(alpha2);# Ersetze NaN durch einen Ausdruck für den zugehörigen Neigungswinkel
e2 = berechne_e(s,alpha2,gamma2);# Ersetze NaN durch einen Ausdruck für die zugehörige Breite der Hilfsfläche e
A2 = berechne_A(e2,b);# Ersetze NaN durch das richtige Argument, sodass A2 der Anteil der Leistung P_M ist, der bei einem Einfallswinkel von 95 Grad am Absorberrohr ankommt 

# Berechne A für alpha = 175 Grad
alpha3 = grad2rad(175);# Ersetze NaN durch einen Ausdruck für den Einfallswinkel (175 Grad) im Bogenmaß
gamma3 = berechne_gamma(alpha3);# Ersetze NaN durch einen Ausdruck für den zugehörigen Neigungswinkel
e3 = berechne_e(s,alpha3,gamma3);# Ersetze NaN durch einen Ausdruck für die zugehörige Breite der Hilfsfläche e
A3 = berechne_A(e3,b);# Ersetze NaN durch das richtige Argument, sodass A3 der Anteil der Leistung P_M ist, der bei einem Einfallswinkel von 175 Grad am Absorberrohr ankommt 

# Prüfung der Ergebnisse (ab hier nichts ändern)
Solarkraftwerk.pruefe_A1A2A3(A1,A2,A3);
```
